```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

from matplotlib.gridspec import GridSpec


import matplotlib as mpl
mpl.rcParams['font.serif'] = "Bitstream Vera Serif"
mpl.rcParams['font.family'] = "serif"
mpl.rcParams['font.size'] = 18
```

```python
silver=np.load('/reg/d/psdm/cxi/cxilr6716/scratch/centerCalib_calibration/silver_run132_z200.npy')
```

```python
def drawArrow(A, B, color):
    plt.arrow(A[0], A[1], B[0] - A[0], B[1] - A[1],
              head_width=0.2, 
              length_includes_head=True,
             color=color, alpha=0.6)
```

```python
centers1 = np.array([[875.9,876.0],
                     [873.8,873.3],
                     [873.8,872.1],
                     [876.4,873.4],
                     [876.3,874.4]])

err1 = np.array([[0.1,0.4],
                 [0.5,1.1],
                 [0.3,0.5],
                 [0.1,0.1],
                 [0.4,0.5]])
z1 = np.array([200.,300.,400.,300.,200.])

centers2 = np.array([[874.0,871.1],
                     [875.9,871.1],
                     [874.6,872.1],
                     [874.8,872.6],
                     [876.0,872.6],
                     [875.3,874.0],
                     [874.1,874.3],
                     [871.2,875.6]])
err2 =np.array([[.7,.5],
                [.4,.8],
                [.3,.3],
                [.3,.3],
                [.4,.9],
                [.6,.3],
                [.3,.7],
                [.7,.3]])

z2 = np.array([400,350,300,250,200,150,100,50])
```

```python
c1 = 'y'
plt.errorbar(centers1[:,0], centers1[:,1], 
             xerr=err1[:,0], yerr=err1[:,1],
            marker='.',ls='none', color=c1,markersize=15)
for ii in range(centers1.shape[0]-1):
    drawArrow(centers1[ii,:],centers1[ii+1,:],'k')
    
for ii in range(centers1.shape[0]):
    plt.text(s=int(z1[ii]),x=centers1[ii,0]-0.1,y=centers1[ii,1]+0.2)
plt.xlabel('Center x position (pixel unit)')
plt.ylabel('Center y position (pixel unit)')

c2 = 'r'
plt.errorbar(centers2[:,0], centers2[:,1], 
             xerr=err2[:,0], yerr=err2[:,1],
            marker='.',ls='none', color=c2,markersize=15)
for ii in range(centers2.shape[0]-1):
    drawArrow(centers2[ii,:],centers2[ii+1,:],'k')
    
for ii in range(centers2.shape[0]):
    plt.text(s=int(z2[ii]),x=centers2[ii,0]-0.1,y=centers2[ii,1]+0.2)
plt.xlabel('Center x position (pixel unit)')
plt.ylabel('Center y position (pixel unit)')
```
