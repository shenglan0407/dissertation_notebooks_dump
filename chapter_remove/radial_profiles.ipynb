```python
def smooth(x, beta=10.0, window_size=11):
    """
    Apply a Kaiser window smoothing convolution.
    
    Parameters
    ----------
    x : ndarray, float
        The array to smooth.
        
    Optional Parameters
    -------------------
    beta : float
        Parameter controlling the strength of the smoothing -- bigger beta 
        results in a smoother function.
    window_size : int
        The size of the Kaiser window to apply, i.e. the number of neighboring
        points used in the smoothing.
        
    Returns
    -------
    smoothed : ndarray, float
        A smoothed version of `x`.
    """
    
    # make sure the window size is odd
    if window_size % 2 == 0:
        window_size += 1
    
    # apply the smoothing function
    s = np.r_[x[window_size-1:0:-1], x, x[-1:-window_size:-1]]
    w = np.kaiser(window_size, beta)
    y = np.convolve( w/w.sum(), s, mode='valid' )
    
    # remove the extra array length convolve adds
    b = (window_size-1) / 2
    smoothed = y[b:len(y)-b]
    
    return smoothed
```

```python
data_dir = '/reg/d/psdm/cxi/cxilp6715/results/combined_tables/finer_q/'
run_file = "run30.tbl"

f = h5py.File(os.path.join(data_dir, run_file), 'r')
print f.keys()
rps=f['radial_profs'].value
```

```python
rp = rps[1000]
beta = 50 # smoothing factor
window_size = 200 # pixel units
order = 250 
smooth_rps=np.zeros_like(rps)
for ii in range(rps.shape[0]):
    smooth_rps[ii] = smooth(rps[ii], beta=beta, window_size=window_size) 
```

```python
ii=1000
plt.figure(figsize=[15,6])
plt.subplot(121)
plt.title('(a)')
plt.plot(rps[ii],'.')
plt.plot(smooth_rps[ii], linewidth=3 )
pk_range=(800, 1000)
plt.vlines(pk_range,0,55, linewidth=2,linestyles='--', alpha=0.6)
plt.ylim(0,55)
plt.xlabel('Radial pixel')
plt.ylabel('Radial Profile (ADU)')

plt.subplot(122)
plt.title('(b)')
plt.imshow(img,aspect='auto', vmax=60,vmin=0, cmap='hot')
plt.colorbar()

# plt.savefig('radial_profile_scatter_shot_examples.png')
```

```python
pk_pos=[]
pk_vals=[]
selected_inds=[]

for ii,rr in enumerate(smooth_rps):
    mx = argrelmax(rr, order=order)[0] 

    # make sure there is only one peak!
    if not len(mx) == 1:
        continue

    # make sure the peak lies in the desired range.. 
    pp=mx[0]
    pv = rr[pp]
    
    
    if not pk_range[0] < pp < pk_range[1] : 
        
        continue

#       make sure the peak value is max in the original profile, 
#       because it was selected using line-subtracted profile

    if not rr[pk_range[0]] < pv and not rr[pk_range[1]] < pv :
        continue 
            
    pk_pos.append(pp)
    pk_vals.append(pv) 
    selected_inds.append(ii)
    
pk_vals =np.array(pk_vals)
pk_pos = np.array(pk_pos)
selected_inds=np.array(selected_inds)
```

```python
plt.figure(figsize=[12,6])
gs = GridSpec(1,2)
gs.update(wspace = 0.3)
plt.subplot(gs[0,0])

plt.title('(a)')
pk_pos_hist=np.histogram(pk_pos, bins=40)
plt.step(pk_pos_hist[1][1:],pk_pos_hist[0],linewidth=5,color='r', alpha=0.6)

plt.xlabel('Peak position (pixel unit)')
plt.ylabel('Number of shots')
plt.xlim(800,920)


plt.subplot(gs[0,1])
plt.title('(b)')
pk_val_hist=np.histogram(pk_vals,bins=40)
plt.step(pk_val_hist[1][1:],pk_val_hist[0],linewidth=5,color='b', alpha=0.6)
plt.xlabel('Peak magnitude (ADU)')
plt.ylabel('Number of shots')
plt.xlim(10,120)

plt.savefig('pk_vals_pos_histograms.png')
```
