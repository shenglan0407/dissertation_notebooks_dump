```python

def normalize(d):
    x=d.copy()
    x-=x.min()
    return x/(x.max()-x.min())
```

```python
all_asym=[]
for kk in range(0,15):
    asym=[asym_kmeans[kk]]
    for ii in range(1,15):
        cc= f_eig['q%d'%qidx]['k%d'%kk]['pca%d'%ii][0,phi_offset:num_phi/2-phi_offset]
        nc=normalize(cc)
        asym.append( (np.abs(nc-nc[::-1])).mean() )
    all_asym.append(asym)
all_asym=np.array(all_asym)
```

```python
plt.plot(all_asym[asym_ind],'o-')
plt.plot(all_asym[sym_ind],'o-')
```

```python
plt.plot(asym_kmeans,all_asym[:,0]-all_asym[:,1],'.')
```

```python
remove=all_asym.argmin(-1)

ave_cor =None
num_shots=0
for kk in range(15):
    if remove[kk]>1:
        cc= f_eig['q%d'%qidx]['k%d'%kk]['pca2'][0,phi_offset:num_phi/2-phi_offset]
#         nc=normalize(cc)
    elif remove[kk]==1:
        cc= f_eig['q%d'%qidx]['k%d'%kk]['pca1'][0,phi_offset:num_phi/2-phi_offset]
#       
    else:
        cc=f['q%d'%qidx]['ave_clus_cor'][kk,phi_offset:num_phi/2-phi_offset]
#         nc=normalize(cc)
    nshots=f['q%d'%qidx]['clus_nshots'][kk]
    if ave_cor is None:
        ave_cor=cc*nshots
    else:
        ave_cor+=cc*nshots
    num_shots+=nshots

ave_cor/=num_shots
```

```python
plt.figure(figsize=[8,5])
###after
y1=normalize(ave_cor)
y2=normalize(f['q%d'%qidx]['ave_cor'][phi_offset:num_phi/2-phi_offset])
plt.plot(exp_cpsi[qidx],y1,'g.',
        linewidth=3,alpha=0.6, label='after',
        color='g', markersize=10)

leg_deg=range(0,16,2)
cc=legfit(exp_cpsi[qidx],y1,leg_deg)
plt.plot(exp_cpsi[qidx],legval(exp_cpsi[qidx],cc), 
         linestyle='-',
         linewidth=5,alpha=0.4,color='g')
plt.xlim(-0.8,0.8)
plt.title('Average correlations')

plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
vpos=[-0.76,-0.44,0.44,0.76,0]
plt.vlines(vpos,0,1.0, linestyle='--', linewidth=6, alpha=0.2)

plt.xlabel('$\cos{\psi}$', fontsize=25)

plt.tight_layout()

plt.savefig('ave_cor_after.png', bbox_in_inches='tight')


nc=normalize(ave_cor)
x = normalize(f['q%d'%qidx]['ave_cor'][phi_offset:num_phi/2-phi_offset])
print (np.abs(nc-nc[::-1])).mean()
print (np.abs(x-x[::-1])).mean()
```
